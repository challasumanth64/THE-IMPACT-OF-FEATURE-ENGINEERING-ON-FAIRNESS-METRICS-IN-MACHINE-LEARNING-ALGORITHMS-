```python
# Install necessary packages
!pip install featuretools fairlearn

# Import libraries
import pandas as pd
import numpy as np
import featuretools as ft
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from fairlearn.metrics import demographic_parity_difference, equalized_odds_difference
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('diabetes.csv')  # Replace with your dataset path if needed

# Display the first few rows of the dataset
df.head()

# Check for missing values
print(df.isnull().sum())

# Basic statistics
df.describe()
```

```python
# Separate features and target variable
X = df.drop(columns=['Outcome'])  # Drop the target column 'Outcome'
y = df['Outcome']  # Define the target variable

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define models
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Naive Bayes': GaussianNB()
}

# Initialize dictionary to store baseline results
baseline_results = {}

# Train each model and calculate performance
for name, model in models.items():
    # Train model
    model.fit(X_train, y_train)

    # Predict and calculate metrics
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Create sensitive feature subset for fairness evaluation
    age_sensitive_test = (df['Age'].iloc[y_test.index] > 50)

    # Fairness evaluation using demographic parity and equalized odds
    demographic_parity = demographic_parity_difference(y_test, y_pred, sensitive_features=age_sensitive_test)
    equalized_odds = equalized_odds_difference(y_test, y_pred, sensitive_features=age_sensitive_test)

    # Store results
    baseline_results[name] = {
        'Accuracy': accuracy,
        'F1 Score': f1,
        'Demographic Parity': demographic_parity,
        'Equalized Odds': equalized_odds
    }

baseline_results

```

```python
import featuretools as ft

# Create an entity set and add the dataframe
df['index'] = df.index  # Adding index column for Featuretools compatibility
es = ft.EntitySet(id="diabetes")
es = es.add_dataframe(dataframe_name="patients", dataframe=df, index="index")

# Apply deep feature synthesis for automatic feature generation
feature_matrix, feature_defs = ft.dfs(entityset=es, target_dataframe_name="patients", max_depth=2)

# Scale the new feature matrix
feature_matrix = feature_matrix.drop(columns=['Outcome'])  # Remove target column from feature matrix
feature_matrix_scaled = scaler.fit_transform(feature_matrix)

# Split feature matrix into train and test sets
X_train_fd, X_test_fd, y_train_fd, y_test_fd = train_test_split(feature_matrix_scaled, y, test_size=0.2, random_state=42)

```

```python
# Initialize dictionary to store feature discovery results
fd_results = {}

# Train each model and calculate performance on new features
for name, model in models.items():
    # Train model
    model.fit(X_train_fd, y_train_fd)

    # Predict and calculate metrics
    y_pred_fd = model.predict(X_test_fd)
    accuracy_fd = accuracy_score(y_test_fd, y_pred_fd)
    f1_fd = f1_score(y_test_fd, y_pred_fd)

    # Create sensitive feature subset for fairness evaluation
    age_sensitive_test_fd = (df['Age'].iloc[y_test_fd.index] > 50)

    # Fairness evaluation
    demographic_parity_fd = demographic_parity_difference(y_test_fd, y_pred_fd, sensitive_features=age_sensitive_test_fd)
    equalized_odds_fd = equalized_odds_difference(y_test_fd, y_pred_fd, sensitive_features=age_sensitive_test_fd)

    # Store results
    fd_results[name] = {
        'Accuracy': accuracy_fd,
        'F1 Score': f1_fd,
        'Demographic Parity': demographic_parity_fd,
        'Equalized Odds': equalized_odds_fd
    }

fd_results
```

```python
# Prepare data for visualization
import matplotlib.pyplot as plt

def plot_metrics(metric_name, baseline_results, fd_results):
    labels = list(baseline_results.keys())
    baseline_values = [baseline_results[model][metric_name] for model in labels]
    fd_values = [fd_results[model][metric_name] for model in labels]

    x = np.arange(len(labels))
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, baseline_values, width, label='Baseline')
    rects2 = ax.bar(x + width/2, fd_values, width, label='Feature Discovery')

    ax.set_ylabel(metric_name)
    ax.set_title(f'{metric_name} by Model and Feature Engineering')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    fig.tight_layout()
    plt.show()

# Plot Accuracy, F1 Score, Demographic Parity, and Equalized Odds for baseline vs. feature discovery
for metric in ['Accuracy', 'F1 Score', 'Demographic Parity', 'Equalized Odds']:
    plot_metrics(metric, baseline_results, fd_results)
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Generate polynomial features
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly.fit_transform(X)  # Apply polynomial features to original features

# Scale the new feature set
X_poly_scaled = scaler.fit_transform(X_poly)

# Split data into train and test sets with new features
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score
from fairlearn.metrics import demographic_parity_difference, equalized_odds_difference

# Define models
all_models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier()
}

# Initialize results dictionary
poly_results = {}

# Filter the sensitive feature to match the test set indices
age_sensitive_test_poly = (df['Age'].iloc[y_test_poly.index] > 50)

# Train each model and calculate performance on polynomial features
for name, model in all_models.items():
    model.fit(X_train_poly, y_train_poly)
    y_pred_poly = model.predict(X_test_poly)

    accuracy_poly = accuracy_score(y_test_poly, y_pred_poly)
    f1_poly = f1_score(y_test_poly, y_pred_poly)

    # Fairness evaluation
    demographic_parity_poly = demographic_parity_difference(y_test_poly, y_pred_poly, sensitive_features=age_sensitive_test_poly)
    equalized_odds_poly = equalized_odds_difference(y_test_poly, y_pred_poly, sensitive_features=age_sensitive_test_poly)

    # Store results
    poly_results[name] = {
        'Accuracy': accuracy_poly,
        'F1 Score': f1_poly,
        'Demographic Parity': demographic_parity_poly,
        'Equalized Odds': equalized_odds_poly
    }

# Display results
poly_results
```
